```python
import pandas as pd

# Load the dataset from the correct URL pointing directly to the CSV file
url = 'https://raw.githubusercontent.com/rizkimahesa/UTS-PDT_Kelompok-Vengeance_AB/main/indian_store.csv'
df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
df = df.drop (columns = ['Product Name','Product Category'])
```

```python
df['Product Price'] = df['Product Price'].apply(lambda x: df['Product Price'].unique().tolist().index(x))
df['Quantity'] = df['Quantity'].apply(lambda x: df['Quantity'].unique().tolist().index(x))
```

```python
df.info()
```

```python
median = df['Product Price'].median()
median_price = df['Product Price (INR)'].median()
median_quantity = df['Quantity'].median()
print("Median", median)
print("Median", median_price)
print("Median", median_quantity)
```

```python
mean = df['Product Price'].mean()
mean_price = df['Product Price (INR)'].mean()
mean_quantity = df['Quantity'].mean()
print("mean", mean)
print("mean price", mean_price)
print("mean quantity", mean_quantity)
```

```python
total_count = len(df)
print("Total data:", total_count)
```

```python
url = 'https://raw.githubusercontent.com/rizkimahesa/UTS-PDT_Kelompok-Vengeance_AB/main/Data%20Kursi%20kereta.csv'
df = pd.read_csv(url)
df
```

```python
kursi_tersedia = df[df['Keterangan'] == 'Tersedia']['No Bangku Kreta'].tolist()
kursi_habis = df[df['Keterangan'] == 'Habis']['No Bangku Kreta'].tolist()
```

```python
jumlah_habis= df['Keterangan'].value_counts()['Habis']
jumlah_tersedia= df['Keterangan'].value_counts()['Tersedia']
```

```python
print("Kursi yang masih tersedia:", kursi_tersedia)
print("Jumlah kursi yang masih tersedia:", jumlah_tersedia)
print("kursi yang habis:", kursi_habis)
print("Jumlah kursi yang habis:", jumlah_habis)
```
